```python
#This function prints table 1 which displays the ID, group score and the overall score
#The function takes the DATAFILE and ABOUT_DATA as input
#Here the empty values of data file are considered to be 0
#If a drop directive is present, then the minimum value of that group is emilinated and the score for the group and the overall score is calculated
#Here G in G1, G2, G3... represents Group

os = [] #list os maintains final overall scores of all students
def table1(DATAFILE, ABOUT_DATA):
    print("Table 1: With Drop Directive")

    a = []
    for i in range(4, len(ABOUT_DATA[0])):   #list a contains all the group numbers mentioned and present in about_data file
        a.append(int(ABOUT_DATA[3][i]))
    num_gp = np.max(a)                       #num_gp holds the largest group number
    gp_nums = np.arange(1, num_gp+1)         #creates an array gp_names containing all the group numbers
    # print(gp_nums)

    gp_names = []
    for u in gp_nums:                        #list gp_names is to contain names of groups like G1 G2...
        gp_names.append("G"+str(u))          #This is to be used for the header of table1
    # print(gp_names)

    print("ID     ",'      '.join(gp_names), "      Overall Score")
    s = []                            #list s is for maintaining the score for each group
    for d in range(1,len(gp_nums)+1): #assigns value to the list s as '1', '2',...
        s.append(str(d))
#     print(s)   
    k = []
    for p in range(4, len(ABOUT_DATA)): #list k contains all the drop directives (Although 1 in this assignment, I have kept it flexible for more) 
        k.append(int(ABOUT_DATA[p][1]))
#     print(k)    
    for i in range(len(DATAFILE)):    #get each student from data file
        for x in gp_nums:             #get each group numbers
            ot = []                   #list ot is created to contain all scores of the group 
            for j in range(3, len(DATAFILE[0])):              #get each score of student i
                weighted_score = (float(DATAFILE[i][j])/float(ABOUT_DATA[2][j+1]))* (float(ABOUT_DATA[1][j+1])) #calculate weighted score
                if(int(ABOUT_DATA[3][j+1]) == x):             #checks if the value belongs to the group being considered in the loop
                    ot.append(weighted_score)                 #adds the score to the list only if it is of the group being considered
            if x in k:                          #checks if the group being considered in the loop is a drop directive
                mini = np.min(ot)               #if it is a drop directive then minimum score of the group is determined
                ot.remove(mini)                 #the minimum score is removed from the list ot
                
            s[x-1] = np.sum(ot)                 #adds all the elements of the list ot which contains the scores of the group to calculate the overall score of the group and assigns it to list s
        
        tot = np.round(np.sum(s), 2)            #calculates final overall score by adding overall scores of all groups and rounds it off to two decimal values
             
        os.append(tot)                          #appends the final overall score of student i to the list os... This list is to be used for table3

        print(DATAFILE[i][0], "  ",'    '.join(map(str, np.round(s,2))), "    ", tot)     #print values
```

```python
#This function prints table 1 which displays the ID, group score and the overall score
#The function takes the DATAFILE and ABOUT_DATA as input
#Here the empty values of data file are considered to be 0
#Drop directive is not considered
#Here G in G1, G2, G3... represents Group


os_without = []        #list os_without maintains final overall scores of all students

def table1_without(DATAFILE, ABOUT_DATA):
    print("Table 1: Without Drop Directive")

    a = []             #list a contains all the group numbers mentioned and present in about_data file
    for i in range(4, len(ABOUT_DATA[0])):
        a.append(int(ABOUT_DATA[3][i]))
    
    num_gp = np.max(a) #determines the largest group number
    gp_nums = np.arange(1, num_gp+1) #creates an array containing all the group numbers
    # print(gp_nums)

    gp_names = []
    for u in gp_nums: #creates a list containing names of groups like G1 G2...
        gp_names.append("G"+str(u))
    # print(gp_names)

    print("ID     ",'      '.join(gp_names), "      Overall Score")

    s = []
    for d in range(1,len(gp_nums)+1): #creates a list for maintaining the score for each group
        s.append(str(d))
    
    for i in range(len(DATAFILE)):  #get each line from data file
        for x in gp_nums:           #get each group numbers
            s[x-1] = 0              #initialize score for group as 0
            for j in range(3, len(DATAFILE[0])):    #get each item of the line
                weighted_score = (float(DATAFILE[i][j])/float(ABOUT_DATA[2][j+1]))* (float(ABOUT_DATA[1][j+1])) #calculate overall score
                if(int(ABOUT_DATA[3][j+1]) == x):   #checks if the value belongs to a particular group
                    s[x-1] += weighted_score        #adds overall score to the respective group score

        tot = np.round(np.sum(s), 2) #calculates final overall score by adding overall scores of all groups
                                     #rounded off to two decimal values 
        os_without.append(tot)       #appends the final overall score of student i to the list os_without... This list is to be used for table3
        print(DATAFILE[i][0], "  ",'    '.join(map(str, np.round(s,2))), "    ", tot)    #print values
```

```python
#This function prints table2 which prints the average, standard deviation, maximum and minimum values for each item
#the function takes DATAFILE1 and ABOUT_DATA as arguments
#The empty values in scores are ignored but the score 0 is kept as it is. 
#Here Avg, STD, Max, Min represents Average, Standard Deviation, Maximum and Minimum respectively

def table2(DATAFILE1, ABOUT_DATA):
    print("Table 2:")
    print("{:<10} {:<10} {:<10} {:<10} {:<10}".format("Type","Avg", "STD", "Max", "Min"))
    for i in range(3, len(DATAFILE1[0])):     #for each item
        l = []                                #list l only maintains values to be considered and ignores the empty ones
        for j in range(len(DATAFILE1)):       #get score of each student for item i 
            if DATAFILE1[j][i] == -1:         #check if the value is empty
                pass                          #ignoes if empty
            else:
                l.append(float(DATAFILE1[j][i]))       #appends to the list l otherwise 
        arr = np.array(l)                              #creates an array containing scores of all students for item i
        a = np.average(arr)                            #finds average 
        stdv = np.std(arr)                             #finds standard deviation
        maxi = np.max(arr)                             #finds maximum value
        mini = np.min(arr)                             #finds minimum value   
        print("{:<10} {:<10} {:<10} {:<10} {:<10}".format(ABOUT_DATA[0][i+1], np.round(a,2), np.round(stdv,2), maxi, mini))  #print values
```
